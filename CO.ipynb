```python
# verify convex combination


import numpy as np

import matplotlib.pyplot as plt


x = np.random.random_integers(10, size=(6,2))
x = np.array([[1,1],[2,2], [1,2], [2,1]])

points = []

for i in range(100):
    for j in range(i):
        points.append((j/100.0) * x[0,:] + ((i/100.0)-(j/100.0)) * x[1,:] + (1-(i/100.0)) * x[2,:])
    
points = np.array(points)

print(points.shape)
z = np.ones(points.shape[0])
plt.scatter(points[:,0], points[:,1], c=z, s=100, cmap=plt.cm.cool, edgecolors='None', alpha=0.75)
plt.colorbar()
plt.show()
```

```python
# back tracking 

import matplotlib.pyplot as plt
import numpy as np
from plotly.tools import FigureFactory as FF

def f(x):
    return np.cos(x)

def df(x):
    return -1 * np.sin(x)

x = np.arange(2, 4.5, 0.1)
y = np.cos(x)
plt.plot(x, y)

# rat x = -2
fx = 2.5 
t = np.arange(0, 1, 0.1)

# f(x) + t * d(f(x)) * d(x)
# d(f(x)) = 2 * x = -4 
# d(x) = postive = 1
# @ x = -2
# 4 + (t * -4 * 1)
print(t + fx)
print(f(fx) + (t * df(fx) * 1))
plt.plot(t + fx, f(fx) + (t * df(fx) * 1))

# alpha = 0.3
a = 0.3
plt.plot(t + fx,  f(fx) + a * (t * df(fx) * 1))
plt.show()
```

```python
# misc

import math
import matplotlib.pyplot as plt
import numpy as np

def normDistance(minD, medD, distance):
    return math.sqrt(min(1, (minD + medD) / (distance + medD)))


minD = np.array([0,1,5,10,30,60,160])
minD = 20
medD = np.array([1,2,3,5,10,15,30,60,100])
medD = [minD + 0]

style = ['.', ',', 'o', 'v', '^', '<', '.', '1','2','3','4']

count = 0
for m in medD:
    distance = np.array([i for i in range(0, 200)])   
    distance += minD
    nd = distance + m
    nd = (minD + m)/nd
    nd = np.minimum(1, nd)
    nd = np.sqrt(nd)
    
    normDistance = nd
    #print(distance, nd)
    
    plt.plot(distance, normDistance, label='{}'.format(m), marker=style[count])
    

    
    count+=1
    
distance = np.array([i for i in range(0, 200)]) 
distance += minD
normDistance = np.array([0.5] * len(distance))
plt.plot(distance, normDistance, label='{}'.format('median'))


                           
plt.legend()
    
plt.show()








```
