```python
import numpy as np
import matplotlib.pyplot as plt


x =  np.linspace(-200,200,2000)
y =  np.linspace(-200,200,200)
X, Y = np.meshgrid(x,y)
%matplotlib inline
```

```python

Z = np.absolute(X)**2 + np.absolute(Y)**2


plt.contour(X, Y, Z)
plt.show()
```

```python
# if x and y are correlated, then error surface is like very elongated eclipse

def squaredError(X, Y):
    # two equations 
    # 4x + 3y = 0
    # 4x + 5y = 0
    
    return (0 - (3*X + 3*Y)) ** 2 + (0 - (3*X + 6*Y))**2


# if we normalize x and y such that they are uncorrelated, either though pca 
# or by making shifting 
def squaredError2(X,Y):
    return (0 - (1*X - 1*Y)) ** 2 + (0 - (1*X + 10*Y))**2

plt.ylim(-200,200)
plt.xlim(-200,200)
CS = plt.contour(X, Y, squaredError(X, Y))

manual_locations = [(-1, -1),(2,2)]
plt.clabel(CS, inline=1, fontsize=10, manual=manual_locations)

plt.show()
plt.contour(X, Y, squaredError2(X, Y))
plt.show()
```
