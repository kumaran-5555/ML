```python
import pandas as pd
import numpy as np
import seaborn as sb

date  = pd.read_csv(r'E:\Git\ML\Kaggle_Bosch\Data\train_date.csv')
date.head()

stations = set([ f.split('_')[1] for f in date.columns.values if 'L' in f])

print(stations)

minmax = pd.DataFrame()




for s in stations:
    print('Computing for ',s)
    fs = [f for f in date.columns.values if s in f]
    
    minmax[s+'_Min'] = date[fs].min(axis=1)
    minmax[s+'_Max'] = date[fs].max(axis=1)
    
    if (minmax[s+'_Max'] - minmax[s+'_Min']).sum() == 0:
        del minmax[s+'_Min']
    else:
        minmax[s+'_Diff'] = minmax[s+'_Max'] - minmax[s+'_Min']
    
    date.drop(fs, axis=1, inplace=True)
        

minmax['Duration'] = minmax.max(axis=1)-minmax.min(axis=1)
minmax['Id'] = date['Id']
minmax.head()


import pickle
pickle.dump(minmax, open(r'E:\Git\ML\Kaggle_Bosch\Data\train_minmax.pkl', 'wb'))

import pandas as pd
import numpy as np
import seaborn as sb

date  = pd.read_csv(r'E:\Git\ML\Kaggle_Bosch\Data\test_date.csv')
date.head()

stations = set([ f.split('_')[1] for f in date.columns.values if 'L' in f])

print(stations)

minmax = pd.DataFrame()




for s in stations:
    print('Computing for ',s)
    fs = [f for f in date.columns.values if s in f]
    
    minmax[s+'_Min'] = date[fs].min(axis=1)
    minmax[s+'_Max'] = date[fs].max(axis=1)
    
    if (minmax[s+'_Max'] - minmax[s+'_Min']).sum() == 0:
        del minmax[s+'_Min']
    else:
        minmax[s+'_Diff'] = minmax[s+'_Max'] - minmax[s+'_Min']
    
    date.drop(fs, axis=1, inplace=True)
        

minmax['Duration'] = minmax.max(axis=1)-minmax.min(axis=1)
minmax['Id'] = date['Id']        
minmax.head()


import pickle
pickle.dump(minmax, open(r'E:\Git\ML\Kaggle_Bosch\Data\test_minmax.pkl', 'wb'))
```

```python
import hashlib

with open(r'E:\Git\ML\Kaggle_Bosch\Data\test_numeric.csv','r') as input:
    with open(r'E:\Git\ML\Kaggle_Bosch\Data\test_numeric.2.csv','w') as output:
        l = input.readline()
        output.write("Hash,"+l)
        for l in input:            
            fields = l[:-1].split(',')[2:]
            hash = hashlib.sha224(','.join(fields).encode('ascii')).hexdigest()
            
            output.write(str(hash) + ','+l)
```
