```python
import pandas as pd
import numpy as np
import seaborn as sb

date  = pd.read_csv(r'E:\Git\ML\Kaggle_Bosch\Data\train_date.csv', nrows=200000)
date.head()
```

```python
stations = set([ f.split('_')[1] for f in date.columns.values if 'L' in f])

print(stations)
```

```python
minmax = pd.DataFrame()

for s in stations:
    fs = date[[f for f in date.columns.values if s in f]]
    
    minmax[s+'_Min'] = fs.min(axis=1)
    minmax[s+'_Max'] = fs.max(axis=1)
    
    if (minmax[s+'_Max'] - minmax[s+'_Min']).sum() == 0:
        del minmax[s+'_Min']
        

minmax['Duration'] = minmax.max(axis=1)-minmax.min(axis=1)
        
minmax.head()
```
