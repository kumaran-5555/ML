# date min max features

```python
import pandas as pd
import numpy as np
#import seaborn as sb

date  = pd.read_csv(r'E:\Git\ML\Kaggle_Bosch\Data\train_date.csv')
date.head()

stations = set([ f.split('_')[1] for f in date.columns.values if 'L' in f])

print(stations)

minmax = pd.DataFrame()




for s in stations:
    print('Computing for ',s)
    fs = [f for f in date.columns.values if '_'+s+'_' in f]
    
    minmax[s+'_Min'] = date[fs].min(axis=1)
    minmax[s+'_Max'] = date[fs].max(axis=1)
    
    if (minmax[s+'_Max'] - minmax[s+'_Min']).sum() == 0:
        del minmax[s+'_Min']
    else:
        minmax[s+'_Diff'] = minmax[s+'_Max'] - minmax[s+'_Min']
    
    date.drop(fs, axis=1, inplace=True)
        
cols = [c for c in minmax.columns.values if c.startswith('S') and '_M' in c]
minmax['all_Min'] = minmax[cols].min(axis=1)
minmax['all_Max'] = minmax[cols].max(axis=1)

minmax['Duration'] = minmax['all_Max'] - minmax['all_Min']
minmax['Id'] = date['Id']
minmax.head()


import pickle
pickle.dump(minmax, open(r'E:\Git\ML\Kaggle_Bosch\Data\train_minmax.2.pkl', 'wb'))

import pandas as pd
import numpy as np
#import seaborn as sb

date  = pd.read_csv(r'E:\Git\ML\Kaggle_Bosch\Data\test_date.csv')
date.head()

stations = set([ f.split('_')[1] for f in date.columns.values if 'L' in f])

print(stations)

minmax = pd.DataFrame()




for s in stations:
    print('Computing for ',s)
    fs = [f for f in date.columns.values if '_'+s+'_' in f]
    
    minmax[s+'_Min'] = date[fs].min(axis=1)
    minmax[s+'_Max'] = date[fs].max(axis=1)
    
    if (minmax[s+'_Max'] - minmax[s+'_Min']).sum() == 0:
        del minmax[s+'_Min']
    else:
        minmax[s+'_Diff'] = minmax[s+'_Max'] - minmax[s+'_Min']
    
    date.drop(fs, axis=1, inplace=True)
        
cols = [c for c in minmax.columns.values if c.startswith('S') and '_M' in c]
minmax['all_Min'] = minmax[cols].min(axis=1)
minmax['all_Max'] = minmax[cols].max(axis=1)

minmax['Duration'] = minmax['all_Max'] - minmax['all_Min']

minmax['Id'] = date['Id']        
minmax.head()


import pickle
pickle.dump(minmax, open(r'E:\Git\ML\Kaggle_Bosch\Data\test_minmax.2.pkl', 'wb'))
```

# Leak features

```python
import pandas as pd
import numpy as np
import seaborn as sb
import pickle

train  = pd.read_csv(r'E:\Git\ML\Kaggle_Bosch\Data\train_date.csv')

train['all_Min'] = train[np.setdiff1d(train.columns, ['Id'])].min(axis=1)
test = pd.read_csv(r'E:\Git\ML\Kaggle_Bosch\Data\test_date.csv')
test['all_Min'] = test[np.setdiff1d(test.columns, ['Id'])].min(axis=1)
test['istest'] = 1
train['istest'] = 0
data = pd.concat((test, train))
del train, test

data = data[['all_Min', 'Id', 'istest']].reset_index(drop=False)

data['IdDiff'] = data.reset_index(drop=True).reset_index(drop=False)['Id'].diff()
data['IdDiffReverse'] = data.reset_index(drop=True).reset_index(drop=False)['Id'][::-1].diff()

data['SortedIdDiff'] = data.sort(['all_Min', 'Id'])['Id'].diff()
data['SortedIdDiffReverse'] = data.sort(['all_Min', 'Id'])['Id'][::-1].diff()

data.head()

data[['all_Min', 'Id', 'SortedIdDiff']].sort(['all_Min', 'Id']).head()

%matplotlib inline
import matplotlib.pyplot as plt

_,ax = plt.subplots(1,1, figsize=(15,4))

data[data['istest']!=2]['SortedIdDiff'].plot.hist(ax=ax, alpha=0.4, color='r', log=True, bins=100)
plt.show()


del data['all_Min']

test = data[data['istest']==1]
del test['istest']

pickle.dump(test, open(r'E:\Git\ML\Kaggle_Bosch\Data\test_leak.pkl', 'wb'))

train = data[data['istest']==0]
del train['istest']

pickle.dump(train, open(r'E:\Git\ML\Kaggle_Bosch\Data\train_leak.pkl', 'wb'))
```
